# Notebook to build a compatible version of rhino3dm from sourcee
Based on https://github.com/mcneel/rhino3dm/blob/main/docs/python/RHINO3DM-BUILD.PY.md.


First, check the runtime's Linux distribution and Python version.

````python
!lsb_release -a
````

````python
!python --version
````

Get dependencies for building `rhino3dm` from source.

````python
!sudo apt install git python3 python3-dev python3-setuptools cmake gcc g++
````

````python
!git clone --recurse-submodules https://github.com/mcneel/rhino3dm.git
````

Switch to the `rhino3dm` directory.

````python
%cd rhino3dm
````

Create the built distribution.

````python
!python setup.py bdist
````

Create a wheel file from distribution.

````python
!python setup.py bdist_wheel
````

At this point, restart the runtime and reset the working directory to root (`/content`).

````python
%cd ..
!ls
````

Use `pip` to install the wheel file.

````python
!pip install /content/rhino3dm/dist/rhino3dm-8.0.0b2-cp310-cp310-linux_x86_64.whl --force-reinstall
````

Test it! Should see:
```
11.0,2.0,3.0
```

````python
from rhino3dm import Point3d, Arc
center = Point3d(1,2,3)
arc = Arc(center, 10, 1)
nc = arc.ToNurbsCurve()
start = nc.PointAtStart
print(start)
````
